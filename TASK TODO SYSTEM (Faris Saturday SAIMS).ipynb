```python
tasktodo = ['assignment' , 'haircut' , 'presentation' , 'dental checkup' ]
taskfinish = []

print("Select your desire action! \n")

signal = True

while signal == True:
    
    ask = input(" Enter 1 to View ToDo List : \n Enter 2 to Add Task to ToDo List : \n Enter 3 to Delete Task From Todo List : \n Enter 4 to Update Task in Todo List : \n Enter 5 to Finish Task from ToDo List : \n Enter 6 to View Finished Task List : \n")
 
    if ask == "1" or "2" or "3" or "4" or "5" or "6":
        
        if ask == "1":
            print("\n Tasks in ToDo List : ", tasktodo)
            sig = input("\n Enter 1 to Go Back : \n Enter 0 to Exit : \n")
            if sig == "1":
                signal = True
            elif sig == "0":
                signal = False
                print("\n System Exit")
    
        elif ask == "2":
            task = input("\n Enter Task to Add : ")
            tasktodo.append(task)
            print("\n New Task Added in ToDo List : ", tasktodo)
            sig = input("\n Enter 1 to Go Back : \n Enter 0 to Exit : \n")
            if sig == "1":
                signal = True
            elif sig == "0":
                signal = False
                print("\n System Exit")
            

        elif ask == "3":
            dtask = input("\n Enter Task to Delete : ")
            if dtask in tasktodo:
                tasktodo.pop(tasktodo.index(dtask))
                print("\n Task Deleted!")
                print("\n ToDo List : ", tasktodo)
                sig = input("\n Enter 1 to Go Back : \n Enter 0 to Exit : \n")
                if sig == "1":
                    signal = True
                elif sig == "0":
                    signal = False
                    print("\n System Exit")
            else:
                print("\n Task not available in ToDo List, Try Again")

        elif ask == "4":
            utask = input("\n Enter Task to Update : ")
            if utask in tasktodo:
                utask2 = input("\n Enter the Updated Task : ")
                val = tasktodo.index(utask)
                tasktodo[val] = utask2
                print("\n Task Updated!")
                print("\n ToDo List : ", tasktodo)
                sig = input("\n Enter 1 to Go Back : \n Enter 0 to Exit : \n")
                if sig == "1":
                    signal = True
                elif sig == "0":
                    signal = False
                    print("\n System Exit")
            else:
                print("\n Task not available in ToDo List, Try Again")
        
        elif ask == "5":
            ftask = input("\n Enter the Task you Want to Finish : ")
            if ftask in tasktodo:
                taskfinish.append(tasktodo.pop(tasktodo.index(ftask)))
                print("\n Task Completed!")
                print("\n ToDo List : ", tasktodo)
                print("\n Finished Task List : ", taskfinish)
                sig = input("\n Enter 1 to Go Back : \n Enter 0 to Exit : \n")
                if sig == "1":
                    signal = True
                elif sig == "0":
                    signal = False
                    print("\n System Exit")
            else:
                print("\n Task not available in ToDo List, Try Again")
        
        elif ask == "6":
            print("\n Tasks Completed : ", taskfinish)
            sig = input("\n Enter 1 to Go Back : \n Enter 0 to Exit : \n")
            if sig == "1":
                signal = True
            elif sig == "0":
                signal = False
                print("\n System Exit")
        
    else:
        print("Please Enter Correct Option \n")
        signal = True

    
```
